## Univariate Linear Regression on the California Housing Prices Dataset

The following task is on the training of a machine learning model based on
a univariate linear regression. The data source is from california housing 
pricess dataset(link to data) [link to data](https://www.kaggle.com/datasets/camnugent/california-housing-prices/download?datasetVersionNumber=1) , having the following features. 

Data Features:

longitude
latitude  
housing_median_age   
total_rooms
total_bedrooms    
population    
households  
median_income
median_house_value


The goal is to run 3 models taking the median_house_value as the dependent variable, predicting the values of the 
media Income, Population and number of households based on the given median_house_value(dependent variable) :

1. Median house values vs Median Income values
2. Median house value vs Population values
3. Median house value vs Household values

```python
import pandas as pd
import plotly.express as px

import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

from sklearn import linear_model

model = linear_model.LinearRegression()


df = pd.read_csv("data/housing.csv")

```

## Data-set description

```python
print(df.describe())

print(df['median_house_value'].describe())

print(df['median_income'].describe())

print(df['population'].describe())

print(df['households'].describe())

df.head()
```

## Seaborn Scatter Plot for Media House values vs Median Income values

```python


sns.scatterplot(x="median_income", y="median_house_value", data=df, color="blue").set_title('Median House value vs. Median Income')
```

## Seaborn Scatter Plot for Media House values vs Population

```python
sns.scatterplot(x="population", y="median_house_value", data=df, color="red").set_title('Median House Value vs. Population')
```

## Seaborn Scatter Plot for Media House values vs Number Of Households

```python
sns.scatterplot(x="households", y="median_house_value", data=df, color="green").set_title('Median House Value vs. Number Of Households')
```

## Plot for the median income against the media house value showing the resulting line from the predictions of the media income values

```python


model.fit(X=df[["median_income"]], y=df[["median_house_value"]])

print(f"Intercept: {model.intercept_}")

vm = model.predict(X=df[["median_income"]])

print(f"mean_absolute_error: {mean_absolute_error(df[['median_house_value']], vm)}")

print(f"mean_square_error: {mean_squared_error(df[['median_house_value']], vm)}")


sns.scatterplot(x="median_income", y="median_house_value", data=df, color="blue").set_title('Median House Value vs. Median Income')

plt.plot(df[["median_income"]], model.predict(df[["median_income"]]), color="black", linewidth=3)

plt.show()
```

## Plot for the population against the media house value showing the resulting line from the predictions of the population values

```python

model.fit(X=df[["population"]], y=df[["median_house_value"]])

print(f"Intercept: {model.intercept_}")

vp = model.predict(X=df[["population"]])

print(f"mean_absolute_error: {mean_absolute_error(df[['median_house_value']], vp)}")

print(f"mean_square_error: {mean_squared_error(df[['median_house_value']], vp)}")


sns.scatterplot(x="population", y="median_house_value", data=df, color="red").set_title('Median House Value vs. Population')

plt.plot(df[["population"]], model.predict(df[["population"]]), color="black", linewidth=3)

plt.show()
```

## Plot for the number of households against the media house value showing the resulting line from the predictions of the  households values

```python


model.fit(X=df[["households"]], y=df[["median_house_value"]])

sns.scatterplot(x="households", y="median_house_value", data=df, color="green").set_title('Median House Value vs. Number Of Households')

plt.plot(df[["households"]], model.predict(df[["households"]]), color="black", linewidth=3)

print(f"Intercept: {model.intercept_}")

vd = model.predict(X=df[["households"]])

print(f"mean_absolute_error: {mean_absolute_error(df[['median_house_value']], vd)}")

print(f"mean_square_error: {mean_squared_error(df[['median_house_value']], vd)}")

      
plt.show()
```
